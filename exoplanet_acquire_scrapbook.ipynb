```python
pd.set_option('display.max_rows', None) 

def column_null_analyzer(df):
    dict_list = []
    for col in df.columns:
        col_nulls = df[col].isnull().sum()
        col_null_percent = col_nulls/len(df.index)
        dict_list.append({'':col,'num_rows_missing':col_nulls,'pct_rows_missing':col_null_percent})
    return pd.DataFrame(dict_list).set_index('').sort_values(by='num_rows_missing', ascending=False)

column_null_analyzer(df)[column_null_analyzer(df).pct_rows_missing < .25]
```

## Refining the query:

```python
url = 'https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+hostname,pl_letter,sy_snum,sy_pnum,cb_flag,discoverymethod,disc_year,glat,glon,pl_orbper,sy_gaiamag,pl_controv_flag,pl_dens,pl_rade,pl_ratdor,pl_masse,st_teff,st_met,st_lum,st_logg,st_age,st_mass,st_dens,st_rad,sy_dist,rowupdate+from+ps&format=csv'
df2 = pd.read_csv(url)
df2.head()
```
